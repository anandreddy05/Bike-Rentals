# Feature Engineering

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/preprocessed-data.csv")
```

```python
for lag in [1,3,5,7]:
    df[f'temp_lag_{lag}'] = df['temp'].shift(lag)
    df[f'atemp_lag_{lag}'] = df['atemp'].shift(lag)
df.head()
```

```python
# Correlation matrix between the lagged value and y
columns = ['y'] + [f"{var}_lag_{lag}" for var in ['temp','atemp'] for lag in [1,3,5,7]]
df[columns].corr()
```

```python
df.dropna(inplace=True)
```

```python
df.to_csv("../data/featured_data.csv")
```
